```python
random.seed(0)
np.random.seed(0)
torch.manual_seed(0)
torch.cuda.manual_seed(0)
torch.backends.cudnn.deterministic = True
```

```python
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader

X_train, X_test, y_train, y_test = train_test_split(
    np.array(data.drop(columns='In-hospital_death')), 
    np.array(data['In-hospital_death']), 
    test_size=0.2, 
    shuffle=True)

X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)
y_train = torch.FloatTensor(y_train).unsqueeze(1)
y_test = torch.FloatTensor(y_test).unsqueeze(1)

batch_size = 8
train_dataloader = DataLoader(Dataset(X_train, y_train), 
                              batch_size=batch_size, 
                              shuffle=True, 
                              num_workers=batch_size)
```

```python
class Dataset():
    def __init__(self, X, Y):
        self.X_data = X
        self.y_data = Y
        
    def __getitem__(self, index):
        return self.X_data[index], self.y_data[index]
    
    def __len__(self):
        return len(self.X_data)
```

```python
class SaveNet(torch.nn.Module):
    def __init__(self, n_hidden_neurons):
        super(SaveNet, self).__init__()
        
        self.fc1 = torch.nn.Linear(32, n_hidden_neurons)  
        self.act1 = torch.nn.ReLU()
        self.bn1 = torch.nn.BatchNorm1d(n_hidden_neurons)
        
        self.fc2 = torch.nn.Linear(n_hidden_neurons, n_hidden_neurons)
        self.act2 = torch.nn.Sigmoid()
        self.bn2 = torch.nn.BatchNorm1d(n_hidden_neurons)
        
        self.dp = torch.nn.Dropout(p=0.1)
        self.fc3 = torch.nn.Linear(n_hidden_neurons, 1) 
        self.act3 = torch.nn.Sigmoid()
        
    def forward(self, x):
        x = self.fc1(x)
        x = self.act1(x)
        x = self.bn1(x)

        x = self.fc2(x)
        x = self.act2(x)
        x = self.bn2(x)
        
        x = self.dp(x)
        x = self.fc3(x)
        x = self.act3(x)
        return x
    
    
net = SaveNet(128)
loss = torch.nn.BCEWithLogitsLoss()
optimizer = torch.optim.RMSprop(net.parameters(), lr=1.0e-3, weight_decay=1e-5)
```

```python
accuracy = lambda preds, y: (preds.round() == y).float().mean().data
mean = lambda l: sum(l) / len(l)

acc_train_hist = []
lss_train_hist = []
acc_test_hist = []
lss_test_hist = []

for epoch in range(100):
    net.train()
    
    acc_train_batch = []
    lss_train_batch = []
    
    for X_batch, y_batch in train_dataloader:
        optimizer.zero_grad()
        preds = net(X_batch)
        loss_val = loss(preds, y_batch)
        loss_val.backward()
        optimizer.step()
        
        lss_train_batch.append(loss_val.data)
        acc_train_batch.append(accuracy(preds, y_batch))
         
    net.eval()
    test_preds = net.forward(X_test)
    
    acc_train_hist.append(mean(acc_train_batch))   
    lss_train_hist.append(mean(lss_train_batch))
    acc_test_hist.append(accuracy(test_preds, y_test))
    lss_test_hist.append(loss(test_preds, y_test).data)
```

```python
from sklearn.metrics import classification_report

net.eval()
preds = net.forward(X_test).round().squeeze().detach().numpy()
test = y_test.squeeze().detach().numpy()
report = classification_report(test, preds)
print(report)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 4))
plt.plot(acc_train_hist, label='accuracy_train')
plt.plot(acc_test_hist, label='accuracy_test')
plt.legend()
```

```python
plt.figure(figsize=(16, 4))
plt.plot(lss_train_hist, label='loss_train')
plt.plot(lss_test_hist, label='loss_test')
plt.legend()
```
