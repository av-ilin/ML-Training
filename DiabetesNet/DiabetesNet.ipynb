```python
random.seed(0)
np.random.seed(0)
torch.manual_seed(0)
torch.cuda.manual_seed(0)
torch.backends.cudnn.deterministic = True
```

```python
class DiabetesNet1(torch.nn.Module):
    def __init__(self, n_hidden_neurons):
        super(DiabetesNet1, self).__init__()
        
        self.fc1 = torch.nn.Linear(8, n_hidden_neurons)
        self.activ1 = torch.nn.Sigmoid()
        self.fc2 = torch.nn.Linear(n_hidden_neurons, n_hidden_neurons)
        self.activ2 = torch.nn.Sigmoid()
        self.fc3 = torch.nn.Linear(n_hidden_neurons, 1)
        self.sigmoid = torch.nn.Sigmoid()
        
    def forward(self, x):
        x = self.fc1(x)
        x = self.activ1(x)
        x = self.fc2(x)
        x = self.activ2(x)
        x = self.fc3(x)
        x = self.sigmoid(x)
        return x
```

```python
class DiabetesNet2(torch.nn.Module):
    def __init__(self, n_hidden_neurons):
        super(DiabetesNet2, self).__init__()
        
        self.layer_1 = torch.nn.Linear(8, n_hidden_neurons) 
        self.layer_out = torch.nn.Linear(n_hidden_neurons, 1) 

        self.relu = torch.nn.ReLU()
        self.sigmoid = torch.nn.Sigmoid()
        self.dropout = torch.nn.Dropout(p=0.1)
        self.batchnorm = torch.nn.BatchNorm1d(n_hidden_neurons)

    def forward(self, x):
        x = self.layer_1(x)
        x = self.relu(x)
        x = self.batchnorm(x)

        x = self.dropout(x)
        x = self.layer_out(x)
        x = self.sigmoid(x)
        return x
```

```python
class DiabetesNet3(torch.nn.Module):
    def __init__(self, n_hidden_neurons):
        super(DiabetesNet3, self).__init__()
        
        self.layer_1 = torch.nn.Linear(8, n_hidden_neurons) 
        self.layer_2 = torch.nn.Linear(n_hidden_neurons, n_hidden_neurons)
        self.layer_out = torch.nn.Linear(n_hidden_neurons, 1) 

        self.relu = torch.nn.ReLU()
        self.sigmoid = torch.nn.Sigmoid()
        self.dropout = torch.nn.Dropout(p=0.1)
        self.batchnorm1 = torch.nn.BatchNorm1d(n_hidden_neurons)
        self.batchnorm2 = torch.nn.BatchNorm1d(n_hidden_neurons)
        
    def forward(self, x):
        x = self.layer_1(x)
        x = self.relu(x)
        x = self.batchnorm1(x)

        x = self.layer_2(x)
        x = self.sigmoid(x)
        x = self.batchnorm2(x)
        
        x = self.dropout(x)
        x = self.layer_out(x)
        x = self.sigmoid(x)
        return x
```

```python
class Dataset():
    def __init__(self, X, Y):
        self.X_data = X
        self.y_data = Y
  
    def __getitem__(self, index):
        return self.X_data[index], self.y_data[index]

    def __len__(self):
        return len(self.X_data)
```

```python
from sklearn.metrics import accuracy_score

def accuracy(preds_raw, y_test):
    preds = torch.round(preds_raw.squeeze()).detach().numpy()
    test = y_test.detach().numpy()
    return accuracy_score(test, preds)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    np.array(data.drop(columns='Outcome')), 
    np.array(data['Outcome']), 
    test_size=0.3, 
    shuffle=True)

X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)
y_train = torch.FloatTensor(y_train)
y_test = torch.FloatTensor(y_test)
```

```python
net = DiabetesNet3(128)
loss = torch.nn.BCEWithLogitsLoss()
optimizer = torch.optim.RMSprop(net.parameters(), lr=1.0e-3, weight_decay=1e-5)
```

```python
dloader = torch.utils.data.DataLoader(dataset=Dataset(X_train, y_train), batch_size=50, shuffle=True)
acc = []
lss = []

for epoch in range(2000):
    acc_batch = []
    loss_batch = []
    for x_batch, y_batch in dloader:
        optimizer.zero_grad()
        preds = net(x_batch)
        loss_value = loss(preds, y_batch.unsqueeze(1))
        loss_value.backward()
        optimizer.step()
        acc_batch.append(accuracy(preds, y_batch))
        loss_batch.append(loss_value)

    acc.append(sum(acc_batch)/len(acc_batch)) 
    lss.append(sum(loss_batch)/len(loss_batch))
```

```python
from sklearn.metrics import classification_report

net.eval()
preds = net.forward(X_test).squeeze().round().detach().numpy()
test = y_test.squeeze().detach().numpy()
report = classification_report(test, preds)
print(report)
```

```python
from matplotlib import pyplot as plt
from sklearn.metrics import roc_curve, precision_recall_curve
auc_roc = roc_curve(test, preds, pos_label=1)
auc_pr = precision_recall_curve(test, preds, pos_label=1)
fig, ax = plt.subplots(1, 3, figsize=(20,4), gridspec_kw={'width_ratios': [1, 1, 3]})

ax[0].plot(auc_roc[0], auc_roc[1], label='roc', color='blue')
ax[0].title.set_text('roc-curve')

ax[1].plot(auc_pr[1], auc_pr[0], label='roc', color='green')
ax[1].title.set_text('auc-curve')

ax[2].plot([i for i in range(len(acc))], acc, color='red', label='accuracy')
ax[2].title.set_text('accuracy')

plt.figure(figsize=(20,4))
plt.plot([i for i in range(len(lss))], torch.Tensor(lss).detach().numpy(), color='red', label='loss')
plt.title('loss')
```
